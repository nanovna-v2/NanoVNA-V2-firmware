# Sample notebook to interact with NanoVNA

```python
cd python
```

```python
%matplotlib inline
from nanovna import NanoVNAV2
#reload script
# %run nanovna.py
```

```python
# create instance and connect to the device
nv = NanoVNAV2()
```

```python
nv.set_sweep(1e6, 300e6)
nv.fetch_frequencies()
```

```python
s11 = nv.data(0)
s12 = nv.data(1)
```

```python
nv.logmag(s11)
nv.logmag(s12)
```

```python
nv.phase(s12)
```

## Various plots by scikit-rf

```python
n = nv.smith(s11)
```

```python
n.plot_s_deg()
```

```python
n.plot_s_db()
```

## read/write touchstone file

```python
n.write_touchstone('100mhz-lpf-s11')
```

```python
import skrf
m = skrf.Network('100mhz-lpf-s11.s1p')
```
